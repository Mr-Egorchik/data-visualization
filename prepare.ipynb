# Подготовка датасета

```python
import pandas as pd
```

```python
df = pd.read_csv('characters.csv')
df.head()
```

```python
df.columns
```

Уберем лишние столбцы

```python
df.info()
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        print(f'Column {col}: {df[col].nunique()} unique values, {df[col].count()}/{df[col].size} not-null values')
```

Убираем столбец 'arkhe', так как им обладает малая часть персонажей

Также уберем столбцы special_{1-5} (оставим только значения 1 уровня и 90 уровня), столбцы voice, ascension_material, ascension_gem, talent_book, talent_material, affiliation, special_dish, constellation

```python
df.drop(['arkhe', 'constellation', 'affiliation', 'special_dish'], axis=1, inplace=True)
```

```python
for i in range(1, 10):
    df.drop(['talent_book_'+str(i)+'-'+str(i+1)], axis=1, inplace=True)
for i in range(1, 6):
    df.drop(['special_'+str(i)], axis=1, inplace=True)
for i in range(3):
    df.drop(['ascension_material_'+str(2*i)+'-'+str(2*i+2)], axis=1, inplace=True)
```

```python
df.drop(['voice_en', 'voice_cn', 'voice_jp', 'voice_kr', 'ascension_gem_0-1', 'ascension_gem_1-3', 'ascension_gem_3-5', 'ascension_gem_5-6', 'talent_material_1-2', 'talent_material_2-6', 'talent_material_6-10'], axis=1, inplace=True)
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        print(f'Column {col}: {df[col].nunique()} unique values, {df[col].count()}/{df[col].size} not-null values')
```

Теперь посмотрим числовые признаки

```python
for col in df.columns:
    if df[col].dtype != 'object':
        print(f'Column {col}: {df[col].count()}/{df[col].size} not-null values')
```

От характеристик оставим только значения 1_20 и 90_90 как параметры на 1 и 90 уровне соответственно

```python
stats = ['hp', 'atk', 'def']
levels = ['20_20', '20_40', '40_40', '40_50', '50_50', '50_60', '60_60', '60_70', '70_70', '70_80', '80_80', '80_90']

for st in stats:
    for lvl in levels:
        df.drop(['_'.join([st, lvl])], axis=1, inplace=True)
```

```python
df.info()
```

```python
df['region'].fillna('Other')
```

```python
df.dropna(subset=['limited'], inplace=True) 
```
